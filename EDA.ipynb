# **EXPLORATORY DATA ANALYSIS**

#### **Distribution of Target Variable [Status]**

```python
sns.countplot(x='Status', data=df)
plt.title('Distribution of default status')
plt.show()


# Calculate the counts of default and non-default loans
default_counts = df['Status'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(default_counts, labels=default_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Default Status Distribution with Percentage')
plt.axis('equal')  
plt.show()
```

#### **Distribution of missing values**

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

#### **Relationship between default status and other variables**

```python
sns.pairplot(df, hue='Status', palette='coolwarm')
plt.title('Pairplot of Features Colored Default Status')
plt.show()
```

#### **Relationship between loan type and default status**

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='loan_type', hue='Status', palette='viridis')
plt.title('Loan Type vs. Default Status')
plt.xlabel('Loan Type')
plt.ylabel('Count')
plt.show()
```

#### **Distribution of loan amount**

```python
sns.displot(df['loan_amount'].dropna(),kde=True,color='darkred',bins=30)
```

#### **Identifying relationship between gender and default status**

```python
fig,ax=plt.subplots()
sns.barplot(data=df,x='Status',y='income',hue='Gender')
fig.set_size_inches([12,6])
plt.show()
```
